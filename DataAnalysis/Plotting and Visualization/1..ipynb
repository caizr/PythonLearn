```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=np.arange(5)
plt.plot(data)
```

```python
#matplotlib图像都位于Figure对象中，用plt.figure创建一个新的Figure
fig = plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
#这时执行一条绘图命令，matplotlib会在最后一个用过的subplot(如果没有则创建一个)
#上进行绘制

#k--是一个线形选项，告诉码头plotlib绘制黑色虚线图
plt.plot(np.random.randn(50).cumsum(),'k--')
```

```python
#上面由fig.add_subplot所返回的对象是  AxesSubplot 对象，直接调用它们的实例
#方法就可以在其他空着的格子里面画图！
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
```

```python
#还有一个更方便的方法plt.subplots，它可以创建⼀个新的Figure，并返回⼀个含有
#已创建的subplot对象的NumPy数组
fig,axes=plt.subplots(2,3)
```

## 刻度、标签和图例

## 设置标题、轴标签、刻度以及刻度标签

```python
#创建一个随机漫步
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
#set_xticks设置模式
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
#set_xticklabels设置模式
f = plt.figure()
ax1=f.add_subplot(1,1,1)
#rotation设定x标签倾斜度30度
labels=ax1.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
#设置标题
ax1.set_title('My plot')
#为x轴设定一个名称
ax1.set_xlabel('Num')
#为y轴设定一个名称
ax1.set_ylabel('A',rotation=40)
```

#### 轴的类有集合的方法，前面的例子同样可以写为

```python
f2=plt.figure()
ax2=f2.add_subplot(1,1,1)

props={'title':'My plot','xlabel':'Num','ylabel':['A']}
ax2.set(**props)
```

## 添加图例

```python
f3=plt.figure();ax4=f3.add_subplot(1,1,1)
ax4.plot(np.random.randn(1000).cumsum(),'k',label='one')
ax4.plot(np.random.randn(1000).cumsum(),'k--',label='two')
ax4.plot(np.random.randn(1000).cumsum(),'b--',label='two')
#legend添加图例,loc指定位置，best会选择合适的位置！
ax4.legend(loc='best')
```

## 图形的绘制

#### 图形的绘制要麻烦⼀些。matplotlib有⼀些表示常⻅图形的对象。这些对象被称为块（patch）。其中有些（如Rectangle和Circle），可以在matplotlib.pyplot中找到，但完整集合位于matplotlib.patches。

```python
f4=plt.figure()
ax5=f4.add_subplot(1,1,1)
#左下角坐标、宽、高、颜色、透明度
rect=plt.Rectangle((0.2,.75),0.4,0.15,color='k',alpha=0.3)
#类似的，还有Circle,Polygon等，更多的在matplotlib.patches库中
ax5.add_patch(rect)
```

## 将图标保存到文件

```python
#保存为svg文件，还可以保存为pdf。
f4.savefig('f.svg')
f4.savefig('f2.png',dpi=400,bbox_inches='tight')
```
