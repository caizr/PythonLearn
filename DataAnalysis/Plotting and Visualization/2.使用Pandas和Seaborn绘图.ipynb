## 柱状图

#### plot.bar()和plot.barn()分别绘制水平和垂直的柱状图。此时，Series和DataFrame的索引将会被用作X(bar)和Y(barn)刻度

```python
#注意这里fig和axes！！
fig,axes=plt.subplots(2,1)
data=pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0],color='k',alpha=0.7)
data.plot.barh(ax=axes[1],color='r',alpha=0.7)
```

#### 对于DataFrame，柱状图将每一行分为一组

```python
#注意pd.Index !!!!!!!
df2=pd.DataFrame(np.random.rand(16).reshape(4,4),index=
                 ['one','two','three','four',
                 ],columns=pd.Index(['A','B','C','D'],name='Genus'))
df2.plot.bar()
```

```python
df2
```

#### 生成堆积柱状图

```python
df2.plot.barh(stacked=True,alpha=0.5)
```

#### 柱状图有一个不错的用法：利用value_counts图形化显示Series中各值出现的频率，比如s.value_counts().plot.bar()

```python
s2=pd.Series([1,3,1,2,4,3,4,3,1,2,1,2,4])
s2.value_counts().plot.bar(color='r',alpha=0.5)
```

## 使用seaborn

```python
tips=pd.read_csv('examples/tips.csv')
```

```python
tips
```

```python
#取某两列，作为新的Pd类型变量
party_counts=pd.crosstab(tips['day'],tips['size'])
party_counts
```

```python
#舍弃size为1和6两列
party_counts=party_counts.loc[:,2:5]
party_counts
```

```python
#对每一行进行归一化处理：
party_pcts=party_counts.div(party_counts.sum(1),axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```
