# 注解及在Subplot上绘图：案例

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)#添加子图
#读取数据，第一列(时间数据)作为索引，关闭时/分/秒
data=pd.read_csv('examples/spx.csv',index_col=0,parse_dates=True)

spx=data['SPX']
#直接用pandas的Series画图
spx.plot(ax=ax1,style='k-')
crisis_data=[
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
#在图上绘图
for date, label in crisis_data:
    ax1.annotate(label, xy=(date, spx.asof(date) + 75),
        xytext=(date, spx.asof(date) + 225),
        arrowprops=dict(facecolor='black', headwidth=4, width=2,
        headlength=4),
        horizontalalignment='left', verticalalignment='top')
    
#放大尺寸
ax1.set_xlim(datetime(2007,1,1),datetime(2011,1,1))
ax1.set_ylim([600,1800])
ax1.set_title('Important dates in the 2008-2009 financial crisis')
```
